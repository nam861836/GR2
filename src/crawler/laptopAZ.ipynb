```python
text_entries = soup.find(class_ = 'col column-2').text.split('\n')
# text_entries
```

```python
# Target text to find
target_text = 'Xem chi tiết cấu hình'

# Find the index of the target text
try:
    target_index = text_entries.index(target_text)
except ValueError:
    print(f"'{target_text}' not found in the list.")
    target_index = None

cleaned_entries = [entry.strip() for entry in text_entries[:target_index] if entry.strip()]
cleaned_entries
```

```python
import re

# Initialize variables to store warranty and stock status
warranty = None
status = None

# Extract warranty and status based on keywords
for entry in cleaned_entries:
    # Clean entry
    entry_cleaned = entry.strip()

    # Extract warranty
    if 'Bảo hành' in entry_cleaned:
        warranty_match = re.search(r'Bảo hành[:\s]*(\d{1,2} tháng|[\w\s]+)', entry_cleaned)
        if warranty_match:
            warranty = warranty_match.group(1).strip()

    # Extract stock status
    if 'Tình trạng' in entry_cleaned:
        status_match = re.search(r'Tình trạng[:\s]*([\w\s]+)', entry_cleaned)
        if status_match:
            status = status_match.group(1).strip()

# Print extracted information
print(f"Warranty: {warranty}")
print(f"Status: {status}")
```

```python
detail = soup.find(class_ = 'table table-striped table-bordered')

# Find all <tr> elements in the detail
rows = detail.find_all('tr')

# Iterate over each row and extract text from each <td>
information = {
    'Name' : name,
    'Price' : price,
    'Deal' : deal,
    'Warranty' : warranty,
    'Status' : status
}
for row in rows:
    cells = row.find_all('td')
    # Extract and print the text from each cell in the row
    row_text = [cell.get_text(strip=True) for cell in cells]
    print(" | ".join(row_text))
    information[row_text[0]] = row_text[1]
```

```python
import requests as re
# link = 'https://cellphones.com.vn/laptop.html'
# link = 'https://laptopaz.vn/laptop-gaming.html'
link = "https://laptopaz.vn/laptop-moi.html"
full_links = [link]
while link is not None:
    resp = re.get(link)
    if resp.status_code ==200:
      soup = BeautifulSoup(resp.content)
      # print("Success!")
    else:
      print("Failed!")
    # Lấy link của trang sau
    links = soup.find_all(class_='page-link')
    for i, tag in enumerate(links):
        if tag.get('href') == '#':
            # Get the next sibling links
            if i + 1 < len(links):
                link = 'https://laptopaz.vn'+links[i + 1].get('href')
                full_links.append(link)
            else:
                link = None
            break

full_links
```

```python
import requests as re

infors = []
for link in full_links:
    print(link)
    # link = 'https://cellphones.com.vn/laptop.html'
    resp = re.get(link)
    if resp.status_code ==200:
      soup = BeautifulSoup(resp.content)
      print("Success!")
    else:
      print("Failed!")


    tag = soup.find_all(class_='p-name')
    # href_values = []
    for i in tag:
      try:
        href_value = 'https://laptopaz.vn'+i.get('href')
        resp = re.get(href_value)
        if resp.status_code ==200:
          soup = BeautifulSoup(resp.content)

        description = soup.find(class_ = 'bg-white border-bottom mb-0 p-3 text-18 title w-100 bk-product-name').text
        condition = description.split(']')[0].strip('[').strip()
        name = description.split(']')[1].split('(')[0].strip()

        text_entries = soup.find(class_ = 'col column-2').text.split('\n')
        # text_entries

        # Target text to find
        target_text = 'Xem chi tiết cấu hình'

        # Find the index of the target text
        try:
            target_index = text_entries.index(target_text)
        except ValueError:
            print(f"'{target_text}' not found in the list.")
            target_index = None

        cleaned_entries = [entry.strip() for entry in text_entries[:target_index] if entry.strip()]


        # Initialize variables to store warranty and stock status
        warranty = None
        status = None

        # Extract warranty and status based on keywords
        for entry in cleaned_entries:
            # Clean entry
            entry_cleaned = entry.strip()

            # Extract warranty
            if 'Bảo hành' in entry_cleaned:
                parts = entry_cleaned.split('Bảo hành:')[1].strip()
                # Extract just the relevant part before any additional text
                warranty = parts.split(' ')[0].strip()

            # Extract stock status
            if 'Tình trạng' in entry_cleaned:
                parts = entry_cleaned.split('Tình trạng:')[1].strip()
                status = parts.strip()

        # # Print extracted information
        # print(f"Warranty: {warranty}")
        # print(f"Status: {status}")

        detail = soup.find(class_ = 'table table-striped table-bordered')

        # Find all <tr> elements in the detail
        rows = detail.find_all('tr')

        # Iterate over each row and extract text from each <td>
        information = {
            'Name' : name,
            'Price' : price,
            'Warranty' : warranty,
            'Status' : status
        }
        for row in rows:
            cells = row.find_all('td')
            # Extract and print the text from each cell in the row
            row_text = [cell.get_text(strip=True) for cell in cells]
            # print(" | ".join(row_text))
            information[row_text[0]] = row_text[1]
        infors.append(information)
      except:
        continue
```

```python
for idx, entry in enumerate(infors):
    if len(entry) != 13:
        print(f"Dictionary at index {idx} has {len(entry)} key-value pairs:")
        print(entry)
```

```python
max_length = 0
dict_with_max_length = None

# Find the dictionary with the most key-value pairs
for entry in infors:
    length = len(entry)
    if length > max_length:
        max_length = length
        dict_with_max_length = entry

# Print the result
print(f"Dictionary with the most key-value pairs ({max_length} pairs):")
print(dict_with_max_length)
```
